```python
#PRIMER REPORTE
import pandas as pd
import sqlite3

# Leer el archivo Excel original
df = pd.read_excel("reactiva.xlsx",header= 1)
# Seleccionar las columnas requeridas
df_ubigeos = df[['UBIGEO', 'REGION', 'PROVINCIA', 'DISTRITO']].drop_duplicates()

# Establecer conexión con la base de datos
conn = sqlite3.connect('ubigeos.db')

# Crear un cursor para ejecutar consultas SQL
cursor = conn.cursor()
# Crear la tabla "ubigeos" si no existe
cursor.execute('''CREATE TABLE IF NOT EXISTS ubigeos (
                    UBIGEO TEXT,
                    REGION TEXT,
                    PROVINCIA TEXT,
                    DISTRITO TEXT
                )''')

# Insertar los datos en la tabla "ubigeos"
for index, row in df_ubigeos.iterrows():
    cursor.execute("INSERT INTO ubigeos (UBIGEO, REGION, PROVINCIA, DISTRITO) VALUES (?, ?, ?, ?)", 
                   (row['UBIGEO'], row['REGION'], row['PROVINCIA'], row['DISTRITO']))

# Confirmar la transacción y cerrar la conexión
conn.commit()
conn.close()
```

```python
#SEGUNDO REPORTE
import pandas as pd

# Leer el archivo Excel
df = pd.read_excel("reactiva.xlsx",header= 1)

df.columns = df.columns.str.lower().str.replace(' ', '_').str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df.columns = df.columns.str.replace('_', ' ')
df = df.rename(columns={'dispositivo 2': 'dispositivo legal', 'estado  ssp': 'estado'})
# Filtrar por obras de tipo Equipamiento Urbano y estados Actos Previos, En Ejecución, Concluido
df_filtered = df[(df['tipologia'] == 'Equipamiento Urbano') & (df['estado'].isin(['Actos Previos', 'En Ejecución', 'Concluido']))]

# Agrupar por región y estado
grouped = df_filtered.groupby(['region', 'estado'])

# Función para obtener el top 5 de costo de inversión
def top_5(df):
    return df.nlargest(5, 'monto de inversion')
# Iterar sobre cada grupo y generar el archivo Excel
for group, data in grouped:
    region, estado = group
    top_5_data = top_5(data)
    if not top_5_data.empty:
        file_name = f"Top5_{region}_{estado.replace(' ', '_')}.xlsx"
        top_5_data.to_excel(file_name, index=False)
        print(f"Archivo '{file_name}' generado con éxito.")
    else:
        print(f"No hay datos para la región '{region}' en el estado '{estado}'. No se generará el reporte.")
```
