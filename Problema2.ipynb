```python
import pandas as pd
df = pd.read_csv('winemag-data-130k-v2.csv')
# Renombrar columnas
df.rename(columns={"Unnamed:0":"ID","country": "nombre", "description": "descripcion", "designation": "denominacion",
"points": "puntuacion", "price": "precio", "province": "provincia",
"region_1": "region_1", "region_2": "region_2", "taster_name": "nombre_catador",
"taster_twitter_handle": "manejador_twitter_catador", "title": "titulo",
 "variety": "variedad", "winery": "bodega"}, inplace=True)
# Crear la columna "rango_precio"
def clasificar_precio(precio):
    if precio <= 20:
        return 'Económico'
    elif precio <= 50:
        return 'Moderado'
    elif precio <= 100:
        return 'Caro'
    else:
        return 'Muy caro'
df['rango_precio'] = df['precio'].apply(clasificar_precio)
import pandas as pd

# Cargar el archivo CSV "paisxcontinente.csv"
paises_continente = pd.read_csv("Paisxcontinente.csv")

# Crear el diccionario de mapeo
mapeo_continente = dict(zip(paises_continente['nombre'], paises_continente['continente']))

# Función para etiquetar el continente
def etiquetar_continente(pais):
    return mapeo_continente.get(pais, "Desconocido")

# Aplicar la función para crear la columna "continente" en tu DataFrame principal
df['continente'] = df['nombre'].apply(etiquetar_continente)
# Reporte 1: Promedio de puntuacion de vinos por continente
promedio_puntuacion_por_continente = df.groupby('continente')['puntuacion'].mean().sort_values(ascending=False)
print("Promedio de puntuacion de vinos por continente:")
print(promedio_puntuacion_por_continente)
# Reporte 2: Precio máximo de vinos por rango de precio
precio_maximo_por_rango = df.groupby('rango_precio')['precio'].max().sort_values(ascending=False)
print("\nPrecio máximo de vinos por rango de precio:")
print(precio_maximo_por_rango)
# Reporte 3: Cantidad de vinos por provincia y su respectivo promedio de precio
cantidad_y_precio_promedio_por_provincia = df.groupby('provincia').agg({'precio': 'mean', 'nombre': 'count'}).sort_values(by='nombre', ascending=False)
cantidad_y_precio_promedio_por_provincia.rename(columns={'nombre': 'cantidad_vinos', 'precio': 'precio_promedio'}, inplace=True)
print("\nCantidad de vinos por provincia y su respectivo promedio de precio:")
print(cantidad_y_precio_promedio_por_provincia.head(10))
# Reporte 4: Distribución de vinos por variedad en cada continente
distribucion_variedad_por_continente = df.groupby(['continente', 'variedad']).size().unstack(fill_value=0)
print("\nDistribución de vinos por variedad en cada continente:")
print(distribucion_variedad_por_continente)
# Guardar el reporte en un archivo CSV
nombre_archivo_csv = "distribucion_variedad_por_continente.csv"
distribucion_variedad_por_continente.to_csv(nombre_archivo_csv)
print(f"El reporte ha sido guardado en {nombre_archivo_csv}")
```
